```python
import h5py

source_path = r"..\tests\240305_PEN_DIP_0001_10_after_gui.h5"
target_path = r'..\tests\240305_PEN_DIP_0001_10_after_gui.h5'
source_group_path = '/entry_0000/data/analysis/frame00000/fitted_peaks'
target_group_path = '/entry_0000/data/analysis/frame00000/detected_peaks'

with h5py.File(source_path, 'r') as f_src, h5py.File(target_path, 'a') as f_dst:
    if target_group_path in f_dst:
        del f_dst[target_group_path]
    f_src.copy(source_group_path, f_dst, target_group_path)
```
